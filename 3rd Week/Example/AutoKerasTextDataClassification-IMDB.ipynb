# Import libraries and data

```python
import os

import numpy as np
import tensorflow as tf
from sklearn.datasets import load_files

import autokeras as ak
```

```python
dataset = tf.keras.utils.get_file(
    fname="aclImdb.tar.gz",
    origin="http://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz",
    extract=True
)

# set path to dataset
IMDB_DATADIR = os.path.join(os.path.dirname(dataset), "aclImdb")

classes = ["pos", "neg"]
train_data = load_files(os.path.join(IMDB_DATADIR, "train"), shuffle=True, categories=classes)
test_data = load_files(os.path.join(IMDB_DATADIR, "test"), shuffle=True, categories=classes)
```

```python
X_train = np.array(train_data.data)
y_train = np.array(train_data.target)
X_test = np.array(test_data.data)
y_test = np.array(test_data.target)

print(X_train.shape)
print(y_train.shape)
print(X_train[0][50])
```

```python
# Initialize the text classifier.
clf = ak.TextClassifier(
    overwrite=True, max_trials=1
) # It only tries 1 model as a quick demo.
# Feed the text classifier with training data.
clf.fit(X_train, y_train, epochs=2)
# Predict with the best model.
predicted_y = clf.predict(X_test)
# Evaluate the best model with testing data.
print(clf.evaluate(X_test, y_test))
```

```python
# Initialize the text classifier.
clf = ak.TextClassifier(
    overwrite=True, max_trials=1
) # It only tries 1 model as a quick demo.
# Feed the text classifier with training data.
split = 5000
X_val = X_train[split:]
y_val = y_train[split:]
X_train = X_train[:split]
y_train = y_train[:split]
clf.fit(X_train, y_train, epochs=2, validation_data=(X_val, y_val))
# Predict with the best model.
predictd_y = clf.predict(X_test)
# Evaluate the best model with testing data.
print(clf.evaluate(X_test, y_test))
```

# AutoKeras Text Data Regression

```python
reg = ak.TextRegressor(overwrite=True, max_trials=1)

split = 5000
X_val = X_train[split:]
y_val = y_train[split:]
X_train = X_train[:split]
y_train = y_train[:split]

reg.fit(X_train, y_train, epochs=2, validation_data=(X_val, y_val))
predicted_y = reg.predict(X_test)
print(reg.evaluate(X_test))
```
