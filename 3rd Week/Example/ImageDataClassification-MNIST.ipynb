# AutoKeras Image Data Classification - MNIST

## Import libraries and data

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import autokeras as ak
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print("X_train.shape :", X_train.shape)
print("y_train.shape :", y_train.shape)
print("y_train[:3] :", y_train[:3])
```

```python
# 클래스의 개수는 총 10개
print(max(y_train), min(y_train))
```

## Image Data Classification

```python
clf = ak.ImageClassifier(overwrite=True, max_trials=1)
clf.fit(X_train, y_train, epochs=5)
```

```python
predicted_y = clf.predict(X_test)
print(predicted_y)
```

```python
print(clf.evaluate(X_test, y_test))
```

```python
clf = ak.ImageClassifier(overwrite=True, max_trials=1)
clf.fit(X_train, y_train, validation_split=0.15, epochs=10)
```

```python
predicted_y = clf.predict(X_test)
print(predicted_y)
```

```python
print(clf.evaluate(X_test, y_test))
```

```python
clf = ak.ImageClassifier(overwrite=True, max_trials=1)
```

```python
split = 50000
X_val = X_train[split:]
y_val = y_train[split:]
X_train = X_train[:split]
y_train = y_train[:split]

clf.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10)
```

```python
predicted_y = clf.predict(X_test)
print(predicted_y)
```

```python
print(clf.evaluate(X_test, y_test))
```
