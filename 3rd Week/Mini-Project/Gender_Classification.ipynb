```python
import numpy as np
import pandas as pd
import sklearn.model_selection
import sklearn.metrics
import sklearn.model_selection

import autosklearn.classification
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/gender_classification_v7.csv')
df.head()
```

```python
df['gender_new'] = np.where(df['gender'] == 'Female', 0, 1)
df
```

```python
df.drop(['gender'], inplace=True, axis=1)
df
```

#### Split train set and test set

```python
X = df.iloc[:, 0:7].to_numpy()
y = df.iloc[:, 7].to_numpy()
```

```python
print(X, y)
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, random_state=42)
```

#### AutoSklearnClassifier

```python
automl = autosklearn.classification.AutoSklearnClassifier(
    time_left_for_this_task=120,
    per_run_time_limit=30
)
automl.fit(X, y, dataset_name='df')
```

```python
print(automl.show_models())
```

```python
predictions = automl.predict(X_test)
print("Accuarcy score:", sklearn.metrics.accuracy_score(y_test, predictions))
```

```python
print(automl.sprint_statistics())
```

#### Description of fit()

```python
from autogluon.tabular import TabularPredictor
```

```python
df
```

```python
subsample_size = 1000
train_data = df.sample(n=subsample_size, random_state=42)

train_data=train_data.sample(frac=1,random_state=100).reset_index(drop=True)
test_data = train_data[750:]
train_data = train_data[:750]
train_data
```

```python
test_data
```

```python
label = 'gender_new'
print("Summary of class variable : \n", train_data[label].describe())
```

```python
save_path = 'agModels-predictClass' # specifies folder to store trained models
predictor = TabularPredictor(label=label, path=save_path).fit(train_data)
```

```python
y_test = test_data[label]
y_test
```

```python
test_data_nolab = test_data.drop(columns=[label], axis=1)
```

```python
test_data_nolab
```

```python
predictor = TabularPredictor.load(save_path) # unnecesary, just demonstrate how to load previously-trained predictor

y_pred = predictor.predict(test_data_nolab)
print("Predictions: \n", y_pred)
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```

```python
predictor.leaderboard(df, silent=True)
```
